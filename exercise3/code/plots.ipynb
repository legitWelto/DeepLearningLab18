```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
file = open("./checkpoints/config_1/testIoU.txt")
conf1 = []
for line in file:
    conf1 += [float(line.split(" ")[1][:-1])]
file.close()

file = open("./checkpoints/config_2/testIoU.txt")
conf2 = []
for line in file:
    conf2 += [float(line.split(" ")[1][:-1])]
file.close()

file = open("./checkpoints/config_3/testIoU.txt")
conf3 = []
for line in file:
    conf3 += [float(line.split(" ")[1][:-1])]
file.close()

file = open("./checkpoints/config_4normed/testIoU.txt")
conf4 = []
for line in file:
    conf4 += [float(line.split(" ")[1][:-1])]
file.close()

epochs = list(range(2,82,2))
fig, (ax) = plt.subplots(1, 1, figsize=(18,12))
ax.set_xlim([2, 80])
ax.set_xlabel('epoch')
ax.set_ylabel('IoU')
ax.plot(epochs, conf1, c='b', label='Config_1')
ax.plot(epochs, conf2, c='r', label='Config_2')
ax.plot(epochs, conf3, c='g', label='Config_3')
ax.plot(epochs, conf4, c='y', label='Config_4')
ax.legend(loc='lower right')
       
fig.canvas.draw()

print("max IoU:")
print("config_1 | config_2 | config_3 | config_4 |")
print("-------------------------------------------")
print("%1.6f | %1.6f | %1.6f | %1.6f |" %(max(conf1), max(conf2), max(conf3), max(conf4)))
```

```python
file = open("./checkpoints/config_4/testIoU.txt")
conf42 = []
for line in file:
    conf42 += [float(line.split(" ")[1][:-1])]
file.close()

file = open("./checkpoints/config_4normed/testIoU.txt")
conf4 = []
for line in file:
    conf4 += [float(line.split(" ")[1][:-1])]
file.close()

epochs = list(range(2,82,2))
fig, (ax) = plt.subplots(1, 1, figsize=(18,12))
ax.set_xlim([2, 80])
ax.set_xlabel('epoch')
ax.set_ylabel('IoU')
ax.plot(epochs, conf42, c='g', label='without Batchnorm')
ax.plot(epochs, conf4, c='y', label='with batchnorm')
ax.legend(loc='lower right')
       
fig.canvas.draw()
```
